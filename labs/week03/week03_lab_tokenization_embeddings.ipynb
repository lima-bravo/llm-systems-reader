## 1. Corpus

Start with a small, human-readable corpus.
You can replace this later with domain-specific text.


```python
corpus = [
    "the system shall record all events",
    "the system shall verify all inputs",
    "the model predicts the next token",
    "verification requires external checks",
    "optimization affects generalization",
]

corpus
```

## 2. Tokenization: word-level baseline

We begin with naive whitespace tokenization to expose its limitations.


```python
def word_tokenize(text):
    return text.lower().split()

tokens = [word_tokenize(s) for s in corpus]
tokens
```

## 3. Vocabulary and indexing


```python
vocab = sorted({tok for sent in tokens for tok in sent})
stoi = {w:i for i,w in enumerate(vocab)}
itos = {i:w for w,i in stoi.items()}

vocab, len(vocab)
```

## 4. Simple embedding model

We train a very small embedding model using a skip-gram-like objective.


```python
class SimpleEmbedding(nn.Module):
    def __init__(self, vocab_size, d_model=16):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, d_model)
        self.proj = nn.Linear(d_model, vocab_size)

    def forward(self, x):
        h = self.emb(x)
        return self.proj(h)

model = SimpleEmbedding(len(vocab)).to(device)
model
```
